```python
# Data streaming for download of a subset of Hugging Face dataset.
# Streaming is recommended with 45GB dataset.
#Images and Metadata.

from datasets import load_dataset
from datetime import datetime
import pandas as pd
import os

os.makedirs("filtered_posters10k", exist_ok=True)

dataset = load_dataset("stzhao/movie_posters_100k_controlnet", split="train", streaming=True)
threshold = datetime(2013, 1, 1)

metadata = []
count = 0

for row in dataset:
    try:
        row_date = datetime.strptime(row["release_date"], "%Y-%m-%d")
        if row_date > threshold and row["image_height"] == 750:
# For images
            img_path = f"filtered_posters/poster_{count}.jpg"
            row["image"].save(img_path, "JPEG", quality=95)
            
# For non-image
            metadata.append({
                "image_path": img_path,
                "title": row["title"],
                "popularity": row["popularity"],
                "release_date": row["release_date"],
                "image_height": row["image_height"],
                "image_width": row["image_width"]
            })
            count += 1
            print(f"Saved {count}: {row['title'][:40]}...")
            
            if count >= 10000:  
                break
                
    except Exception:
        continue


df = pd.DataFrame(metadata)
df.to_csv("movies_10y_metadata10k.csv", index=False)
```
